```python
import pandas as pa

df = pa.read_csv('Sport car price.csv')

df.head(5)
```

```python
# delete none values
df = df.dropna()

# delete duplicate values
df = df.drop_duplicates()
```

```python
df.sample(20)
```

```python
df.dtypes
```

```python
def convert_to_float(value):
    try:
        value = value.split(' ')[0]
        return float(value)
    except ValueError:
        return 0.0
    
def convert_to_int(value):
    try:
        value = value.replace(',','')
        return int(value)
    except ValueError:
        return 0
```

```python
df['Engine Size (L)'] = df['Engine Size (L)'].apply(convert_to_float)
df['0-60 MPH Time (seconds)'] = df['0-60 MPH Time (seconds)'].apply(convert_to_float)

df['Horsepower'] = df['Horsepower'].apply(convert_to_int)
df['Torque (lb-ft)'] = df['Torque (lb-ft)'].apply(convert_to_int)
df['Price (in USD)'] = df['Price (in USD)'].apply(convert_to_int)
```

```python
summary_stats = pd.DataFrame(columns=df.columns)

summary_stats = summary_stats.drop(['Car Make', 'Car Model'], axis=1)
cols = summary_stats.columns
summary_stats.insert(0, "stats", ['mean', 'median', 'mode', 'std', 'range'], True)
```

```python
summary_stats
```

```python
df['Horsepower'].mode()[0]
```

```python
for col in cols:
    summary_stats.loc[0, col] = df[col].mean()
    summary_stats.loc[1, col] = df[col].median()
    summary_stats.loc[2, col] = df[col].mode()[0]
    summary_stats.loc[3, col] = df[col].std()
    summary_stats.loc[4, col] = df[col].max() - df[col].min()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Horsepower']]
y = df['Price (in USD)']

model = LinearRegression()
model.fit(X, y)

df['Predicted_Price'] = model.predict(X)

plt.plot(df['Horsepower'], df['Predicted_Price'], color='red', label='Linear Regression Line')
plt.scatter(X, y, s=2, color='blue', label='Scatter points')

plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Price vs Horsepower with Linear Regression Line')
plt.legend()

plt.show()
```

```python
x = df['0-60 MPH Time (seconds)']
plt.figure(figsize=(10, 6))
plt.hist(x, bins=[i * 0.5 for i in range(int(x.max() / 0.5) + 1)], edgecolor='black')

plt.xlabel('0-60 MPH Time (seconds)')
plt.ylabel('Frequency')
plt.title('Histogram of 0-60 MPH Times with 0.5 Second Bins')
plt.grid(True)

plt.show()
```
